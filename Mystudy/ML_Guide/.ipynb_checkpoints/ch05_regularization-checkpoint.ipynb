# Regularization

  
회귀 모델에 있어서 가중치들을 전반적으로 감소시키는 효과를 가짐.  
하지만 l1의 경우 미분이 불가능하기 때문에 gradient 기반의 모델에서 주의해야함

# Ridge

$$cost = RSS(W) + \alpha||w||_2^2$$

위와 같이 linear regression에서 regularization term으로 l2를 사용한 형태

# Lasso

$$cost = RSS(W) + \alpha||w||_1$$

위와 같이 linear regression에서 regularization term으로 l1 사용

# Elastic Net

$$cost = RSS(W) + \alpha||w||_2^2 + \beta||w||_1$$

regularization term으로 l1, l2 모두 사용
